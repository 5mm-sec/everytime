```python
driver.find_element_by_xpath('//*[@id="container"]/ul/li[1]').click()

pre_count = 0


while True:

    element = driver.find_elements_by_css_selector("#subjects > div.list > table > tbody > tr")


    result = element[-1]

    driver.execute_script('arguments[0].scrollIntoView(true);',result)
    sleep(2)


    current_count = len(element)
    if pre_count == current_count:
        break

    pre_count = current_count
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
```

```python
trs = soup.select('#subjects > div.list > table > tbody > tr')
results = []
```

```python
results = []
for tr in trs:
    try:
        result=[]
        tds = tr.select('#subjects > div.list > table > tbody > tr > td')
        result.append(tds[1].text) #과목명
        result.append(tds[2].text) #교수
        result.append(tds[6].text) #학년
        result.append(tds[11].text) #수강대상

        sample = tds[8].find('a') # 과목 고유 번호
        if(float(sample['title']))<=0:
            continue

        subject_code =  sample['href'].replace('/lecture/view/','') # 평점 평균
        rating = sample['title']
        result.append(subject_code)
        result.append(rating)
        
        results.append(result)
    except:
        pass
    
print(results)

if results:
    print("성공!!")

```

```python
r2 = []
for r in results:
    r2.append("_".join(r))
    
r3 = list(set(r2))
results = []
for i in r3:
    results.append(i.split("_"))
from datetime import datetime
now = datetime.now()
ns =  now.strftime('%Y_%m_%d_%H_%M_%S')
ns = str(ns)
print(ns)
excel_column = 6
write_wb = Workbook()
write_ws = write_wb.create_sheet(ns +'.xls')
for data in results:
    write_ws = write_wb.active
    write_ws.append(data)
    
    
write_wb.save('C:/Users/cslab/Desktop/selenium/' + ns +'.csv')
```

```python
from selenium import webdriver

import csv
import requests
from bs4 import BeautifulSoup
from time import sleep
from openpyxl import Workbook,load_workbook

driver = webdriver.Chrome(executable_path=r'D:\chromedriver.exe')
url = 'https://everytime.kr/lecture/view/'
results = []
with open('C:/Users/cslab/Desktop/selenium/2022_07_30_19_19_25.csv','rt',encoding='UTF8')as f:
    reader = csv.DictReader(f)
    sb_codes = []
    for row in reader:
        sb_codes.append(row['과목코드'])


for sb_code in sb_codes:
    

    driver.get('https://everytime.kr/login')

    driver.implicitly_wait(5)
    
    sleep(4)


    driver.find_element_by_name('userid').send_keys('')
    driver.find_element_by_name('password').send_keys('')
    sleep(4)

    driver.find_element_by_xpath('//*[@id="container"]/form/p[3]/input').click()
    
    link = url + f'{sb_code}'

    driver.get(link)
    sleep(4)
    
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
     
    head = soup.select('#container > div.side.head')
    trs = soup.select('#container > div.side.article > div.articles > article')
    
    
    articles = driver
    a = 1
    
    for n in trs:
        result = []
        head = driver.find_element_by_xpath('//*[@id="container"]/div[2]/h2').text
        result.append(head)
        prof = driver.find_element_by_xpath('//*[@id="container"]/div[2]/p[1]/span').text
        result.append(prof)
        score = driver.find_element_by_xpath('//*[@id="container"]/div[4]/div[2]/article[1]/p[1]/span/span')
        score1 = score.get_attribute('style').replace('width: ','')
        score2 = score1.replace('%;', '')
        score3 = int(score2) / 20
        result.append(score3)
        sub = driver.find_element_by_xpath('//*[@id="container"]/div[4]/div[2]/article[{}]/p[3]'.format(a)).text
        a += 1
        result.append(sub)        
        results.append(result)
    print(results)
    print('\n')
    if results:
        print("성공")
        
        
excel_column = 4
write_wb = Workbook()
write_ws = write_wb.create_sheet('result8.xls')
for data in results:
    write_ws = write_wb.active
    write_ws.append(data)
write_wb.save('C:/Users/cslab/Desktop/selenium/result8.csv')

```
